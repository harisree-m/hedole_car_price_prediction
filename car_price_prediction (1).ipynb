The `groups()` method returns a tuple that contains all the subgroups of the match. It can return any number of groups that are in a pattern. If there is no group in the pattern, then it returns the default value, which is None.


A groupby operation involves some combination of splitting the object, applying a function, and combining the results. This can be used to group large amounts of data and compute operations on these groups. Used to determine the groups for the groupby.

```python
petrol_data=car_data.groupby('Fuel_Type').get_group('Petrol')
petrol_data.describe()
```

```python
seller_data=car_data.groupby('Seller_Type').get_group('Dealer')
seller_data.describe()
```

```python
car_data.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)
car_data=pd.get_dummies(car_data,columns=['Seller_Type','Transmission'],drop_first=True)
```

```python
X=car_data.drop(['Car_Name','Selling_Price'],axis=1)
y=car_data['Selling_Price']
```

```python
print("Shape of x is:",X.shape)
print("Shape of y is:",y.shape)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
print("X_test shape=",X_test.shape)
print("X-train shape",X_train.shape)
print("y_test shape",y_test.shape)
print("y_train shape",y_train.shape)
```

```python
scaler =StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python

print("MAE:",(metrics.mean_absolute_error(pred,y_test)))
print("MAE:",(metrics.mean_squared_error(pred,y_test)))
print("TR2 score :",(metrics.r2_score(pred,y_test)))
```

```python
sns.regplot(x=pred,y=y_test)
plt.xlabel("Predicted Price")
plt.ylabel("Actual Price")
plt.title("Actual vs Predicted price")
plt.show()
```
